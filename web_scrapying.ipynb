```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://lyrics.az/britney-spears/allsongs.html'
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
response = requests.get(url,headers=headers)
response
```

```python
soup = BeautifulSoup(response.text,'html.parser')
soup
```

```python
soup.head.title
```

```python
soup.find('div').find('td')
```

```python
song=soup.find('div').find('td').a['href']
song
```

```python
type(song)
```

```python
song_list=soup.find_all(class_= 'px-2 py-1 default-link list-group-item-action d-block')
type(song_list)
song_list
```

```python
song_links=[]

for item in song_list:
    try:
        if item is not None:
            song_links.append(item['href'])            
    except:
        continue
song_links
```

```python
song_titles=[]
for item in song_list:
    try:
        if item is not None:
            song_titles.append(item.text)            
    except AttributeError:
        continue
```

```python
song_titles
```

```python
import pandas as pd
df=pd.DataFrame(song_links,song_titles)
df
```

```python
df['lyric_text']='blank'
df
```

```python
df = df.rename(columns={0:'links','lyric_text':'lyrics text'})
df
```
